### Preprocess the Input Data

```python
train_images.shape
```

```python
# Prepare the data into a format the model expects it to be
#-------------------------------------------------------------

# Preprocess Training data
# we will keep the 3D shape of each image as it is, since we will be using Convolution layers 
#..in the begenning. No flattening
# But we convert the data type into float (astype) from integer and scale values between 0 and 1 
#..by dividing with 255

train_images = train_images.astype("float32") / 255 
# Convert into a float32 with values scaled between [0,1] instead of [0,255]

#Preprocess Test data
test_images = test_images.astype("float32") / 255 
# Convert into a float32 with values scaled between [0,1] instead of [0,255]
```

### Build the Neural Network Model Architecture

```python
# See sample architecture we discussed earlier.  
# We will make changes to thisarchitecture by adding more layers and filters
```

<img style="float: left;" src="./images/cnn2.png" width="1000">

```python
# Set random seeds for reproducibility
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# Build the model's neural network architecture with CNN's using functional API
# We will add more layers than the above schematic and also more filters

def model_cifar(): 
    inputs = keras.Input(shape = (32,32,3)) # Define Input shape

    x = layers.Conv2D(filters=32, kernel_size = 3, activation="relu")(inputs) 
    # Convolution Layer with no padding and stride=1 (default)
    
    x = layers.MaxPooling2D(pool_size=2, strides = (2,2))(x) 
    # MaxPool Layer with size = 2 x 2, strides = 2

    x = layers.Conv2D(filters=64, kernel_size = 3, activation="relu")(x) 
    # Convolution Layer with no padding and stride=1 (default)
    
    x = layers.MaxPooling2D(pool_size=2, strides = (2,2))(x) 
    # MaxPool Layer with size = 2 x 2, strides = 2

    x = layers.Conv2D(filters=128, kernel_size=3, activation="relu")(x) 
    # Convolution Layer with no padding and stride=1 (default)

    x = layers.Flatten()(x) # Flatten

    outputs = layers.Dense(10, activation="softmax")(x) # Dense output Layer
    
    model = keras.Model(inputs=inputs, outputs=outputs)
    return model
```

```python
model = model_cifar()
model.summary()
```

### Compile & Train the Model

```python
#Compile the Model by configuring the loss function, optimizer type, & metrics to monitor the model performance

model = model_cifar()


sgd = tf.keras.optimizers.SGD(lr=0.01, momentum=0.9) # we will use sgd optimizer

model.compile(optimizer=sgd, loss='sparse_categorical_crossentropy',  metrics = ["accuracy"])

def lr_scheduler(epoch):
    return 0.01 * (0.5 ** (epoch // 20))

reduce_lr = keras.callbacks.LearningRateScheduler(lr_scheduler)

model_checkpoint = keras.callbacks.ModelCheckpoint(filepath = "./models/model_cifar.keras",save_best_only=True, monitor="val_loss")

callbacks = [model_checkpoint, reduce_lr]

history = model.fit(train_images, train_labels, epochs = 30, batch_size = 128, 
                    validation_split = 0.2, callbacks=callbacks)
```

### Plotting the loss & Accuracy curves

```python
plot_history(history)
```

```python
# Model started overfitting after around 10th epoch
```

### Evaluate the trained model on previously unseen test data 

```python
model.evaluate(test_images, test_labels) 
```
