## 2.  Univariate Time Series Forecasting

```python
# We will first only use the the East side data to predict its own time series data
# input_features = 1 : for univariate timeseries analysis, > 1 for multivariate analysis
```

```python
train_data['East'].shape
```

```python
train_data['East'].head(31)
```

```python
# Inputs = We will use previous 30 days of data to predict the total bicycle count for future. 
# ..so our Sequences will be of length 30.

# Targets = we will forecast the bicyle count for next day

# so basically we will reorganize the training data into sequences of length 30, and 
# ...target would be the next day's count.
```

**keras.utils.timeseries_dataset_from_array**

```python
sequence_length = 30
#we will predict next days bicycle count using past 30 days of data.

# Keras expects the data to be numpy arrays
train = keras.utils.timeseries_dataset_from_array(data = train_data['East'].to_numpy(),
                                                     targets = train_data['East'][sequence_length:],
                                                     sequence_length = sequence_length,
                                                     batch_size=32,
                                                     seed=0
                                                    )
val =  keras.utils.timeseries_dataset_from_array(data = val_data['East'].to_numpy(),
                                                     targets = val_data['East'][sequence_length:],
                                                     sequence_length = sequence_length,
                                                     batch_size=32,
                                                     seed=0
                                                    )
test = keras.utils.timeseries_dataset_from_array(data = test_data['East'].to_numpy(),
                                                     targets = test_data['East'][sequence_length:],
                                                     sequence_length = sequence_length,
                                                     batch_size=32,
                                                     seed=0
                                                    )
```

## Build Neural Network Architectures

### Simple  Dense Model

```python
display(Image("./images/dense.jpg" , width=200))
```

```python
def model_simple_dense(): 
    inputs = keras.Input(shape = (30,)) # Define Input shape  
    outputs = layers.Dense(1)(inputs) # Dense output Layer
    
    model = keras.Model(inputs=inputs, outputs=outputs)
    return model
```

```python
model = model_simple_dense()
model.summary()
```

```python
path = Path("./models/model_simple_dense.keras")
```

```python
# Let's create a function to train and evaluate for convenience

def train_evaluate(model,path,train,val):
    
    # CALL BACKS
    #------------
    # Save only best model
    checkpoint_cb = keras.callbacks.ModelCheckpoint(filepath = path,
                                                       save_best_only=True, monitor="val_loss") 
    
    # Early stop call back
    earlystop_cb = keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, 
                                                 restore_best_weights=True)
    
    callbacks = [checkpoint_cb,earlystop_cb]

    # OPTIMIZER
    # -------------
    sgd = tf.keras.optimizers.SGD(learning_rate=0.02, momentum=0.9)

    # COMPILE
    #---------
    model.compile(optimizer=sgd, loss='huber',  metrics = ["mae"])
    
    # Train the model
    #----------------
    history = model.fit(train, validation_data = val, callbacks=callbacks, epochs=500)
    
    # Evaluate the model on validation data
    #--------------------
    val_loss, val_mae = model.evaluate(val)
    
    return (history,val_mae*3000) # multiplying val_mae with 3000, since we scaled the data earlier
```

```python
(history_simple_dense,val_mae_simple_dense) = train_evaluate(model,path,train,val)
```

```python
plot_history(history_simple_dense,"mae")
```

**MAE with the Simple Dense Model**

```python
# MAE for the baseline forecasing on Validation Dataset was ~ 324

# Mean absolute error with this simple 1 neuron dense model 
print (f"Mean Absolute Error on Validation Dataset is {val_mae_simple_dense}")
```

### Simple RNN model with just 1 Neuron

```python
# Recurrent layers in Keras expects 3D inputs of shape (batch_size, time_steps, input_features)

# batch size is automatically inferred at the time of training

# input_features = 1 : for univariate timeseries analysis, > 1 for multivariate analysis
```

```python
def model_simple_rnn(): 
    inputs = keras.Input(shape = (30,1)) # Define Input shape (timesteps, input_features). 
    outputs = keras.layers.SimpleRNN(1)(inputs) # Dense output Layer
    
    #outputs = keras.layers.Dense(1)(x)
    
    model = keras.Model(inputs=inputs, outputs=outputs)
    return model
```

### Simple RNN model with 16 neurons

```python
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)
```

```python
display(Image("./images/simple_rnn.jpg" , width=800))
```

```python
def model_rnn_16(): 
    inputs = keras.Input(shape = (30,1)) # Define Input shape  
    x = keras.layers.SimpleRNN(16)(inputs) # RNN Layer with 16 neurons
    x = layers.LayerNormalization()(x) 
    
    outputs = keras.layers.Dense(1)(x) # Final Dense layer with just 1 unit
    
    
    model = keras.Model(inputs=inputs, outputs=outputs)
    return model
```

```python
model = model_rnn_16()
model.summary()
```

### Deep RNN

```python
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)
```

```python
display(Image("./images/deep_rnn.jpg", width=800))
```

```python
def model_deep_rnn(): 
    inputs = keras.Input(shape = (30,1)) # Define Input shape  
    x = keras.layers.SimpleRNN(16, return_sequences=True)(inputs) # RNN Layer with 16 neurons
    x = keras.layers.SimpleRNN(16, return_sequences=True)(x) # RNN Layer with 16 neurons
    x = keras.layers.SimpleRNN(16)(x) # RNN Layer with 16 neurons
    
    outputs = keras.layers.Dense(1)(x) # Final Dense layer with just 1 unit
    
    
    model = keras.Model(inputs=inputs, outputs=outputs)
    return model
```

```python
model = model_deep_rnn()
model.summary()
```

```python
path = Path("./models/model_deep_rnn.keras") # name the path the save the model after training

(history_deep_rnn,val_mae_deep_rnn) = train_evaluate(model,path,train,val)
```
